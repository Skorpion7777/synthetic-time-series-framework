```python
new_time_series = time_series[[187]].copy()
new_time_series.insert(0, 'ID', range(1, 1 + len(new_time_series)))
new_time_series.head()
```

```python
labels = pd.read_csv("Dataset/Anonymized/ooout_10_10.csv",header = None)
labels.head()
```

```python
# Rename column 0 to ID in labels
labels.rename(columns={0: 'ID'}, inplace=True)
new_time_series.rename(columns={187: 'Label'}, inplace=True)

# Merge labels with new_time_series based on ID
merged_df = pd.merge(new_time_series, labels, on='ID')

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
null_values = new_time_series['Label'].isnull().sum()
print(f"Number of null values in 'Label' column: {null_values}")
```

```python
merged_df.pop(188)
merged_df.pop(189)
merged_df.rename(columns={'label': 187}, inplace=True)
merged_df.head()
```

```python
# Function to calculate the median of an interval string
def interval_median(interval_str):
    lower, upper = map(float, interval_str.strip('[]').split('-'))
    return (lower + upper) / 2

# Apply the function to columns 1 to 187
for col in range(1, 188):
    merged_df[col] = merged_df[col].apply(interval_median)

# Display the first few rows of the updated dataframe
merged_df.head()
```

```python
label_counts = merged_df['Label'].value_counts()
print(label_counts)
```
